```python
import json
import jsonschema
```

```python
ratingSchema = json.load(open("persistence/schemas/rating-schema.json"))
```

```python
ratingSchema
```

```python
rating = {
    'id': 0,
    'user': "aman",
    'timestamp': "2022-03-29T19:29:28+00:00",
    'artist': "Frank Ocean",
    'baddie': 80,
    'banger': 100
}
```

```python
try:
    jsonschema.validate(instance=rating, schema=ratingSchema)
except jsonschema.exceptions.ValidationError as err:
    print(err)
```
